```python
# tf-idf model
import gensim
from gensim import models
import numpy as np
from gensim.corpora import Dictionary
from gensim.test.utils import get_tmpfile

# Target document to score
documents = ["the dog runs fast"]
line = documents[0]

# Get previously created dictionary
textsDictionary = Dictionary.load_from_text('../data/texts_dict.txt')
uniqueWordsList = list(set(line.split()))
tfidfList = [[textsDictionary[id], freq, textsDictionary.dfs[id]] for id, freq in textsDictionary.doc2bow(simple_preprocess(line))]
wordsinDict = [i[0] for i in tfidfList]


for term in uniqueWordsList:
    if term not in wordsinDict:
        count = 0
        for word in line.split():
            if word == term:
                count += 1
        tfidfList.append([term, count, 1])
 
tfidfCalcList = []

total = 0
for x in tfidfList:
    tfidfTerm = -1*x[1]*np.log(x[2]/len(textsDictionary))
    total += tfidfTerm
    tfidfCalcList.append([x[0], np.around(tfidfTerm, 3)])

for x in tfidfCalcList:
    x[1] = np.around(x[1]/total, 3)

print(tfidfCalcList)
#print(np.log(1/(3.46892*10**-5)))
#print(tfidfList)
#print(len(textsDictionary))
#print(textsDictionary)
# Now we gotta estimate this for n-grams
```

```python
import textblob
import numpy
from smart_open import smart_open

data = [[x for x in line.split("\n") if x != ''] for line in open('../data/sampletweets.txt', encoding='utf-8')]
tweets = [x for x in data if x != []]


tweetDict = {}
emojis = []
tweet = []
tweetScores = []

count = 0
# Get the emojis
for i in tweets:
    if count%11 == 0:
        emojis.append(i[0][-1])
    else:
        tweet.append(i[0])
    count += 1

for x in tweet:
    blob = textblob.TextBlob(x)
    tweetScores.append(blob.sentiment)
    
for x in range(len(emojis)):
    polarity = []
    subjectivity = []
    for y in range(10*x, 10*x+10):
        polarity.append(tweetScores[y].polarity)
        subjectivity.append(tweetScores[y].subjectivity)
    #tweetDict[emojis[x]] = [numpy.mean(polarity), numpy.std(polarity), numpy.mean(subjectivity), numpy.std(subjectivity)]
    tweetDict[emojis[x]] = ["Polarity mean: " + str(numpy.around(numpy.mean(polarity), 3)), "Polarity StDev: " + str(numpy.around(numpy.std(polarity), 3)), "Subjectivity mean: " + str(numpy.around(numpy.mean(subjectivity), 3)), "Subjectivity StDev: " + str(numpy.around(numpy.std(subjectivity), 3))]

for keys,values in tweetDict.items():
    print(keys, values)
```
